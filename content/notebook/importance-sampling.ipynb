Importance Sampling
===================
Importance sampling is a technique for estimating the expectation of a random variable $f(x)$ under distribution $p$ from samples take from a different distribution $q$.

$$
\mu \dot= \mathbb{E}_{x \sim p} \left[ f(x) \right]
$$

Importance sampling is based on the observation that $\mu$ is equivalently expressed as the expectation of a different random variable $f^*(x)=\frac{p(x)}{q(x)}\! \cdot\! f(x)$ under $q$.

$$
\mu = \mathbb{E}_{x \sim q} \left[ \frac{p(x)}{q(x)} f(x) \right]
$$



```python
import pylab as pl
import numpy as np
from numpy import log, exp, sqrt, array
from scipy.stats import norm, uniform, laplace, norm

class mixture(object):
    "Create a mixture distribution out of other distributions."
    def __init__(self, coeff, dists):
        coeff = array(coeff)
        assert coeff.sum() == 1 and (coeff >= 0.0).all(), 'coeff should be a convex combination'
        self.coeff = coeff
        self.dists = dists

    def pdf(self, x):
        return self.coeff.dot([d.pdf(x) for d in self.dists])

    def rvs(self, size):
        "generate samples"
        samples = []
        # pick mixture component with probability proportional to coeff 
        r = np.random.uniform(size=size)
        z = np.cumsum(self.coeff).searchsorted(r)
        for i in xrange(size):
            # then call its rvs method
            samples.append(self.dists[z[i]].rvs())
        return np.array(samples)
```

Let's invent a complicated distribution $p$, which is a mixture of Laplace distributions.

```python
p = mixture([0.5, 0.5],
            [norm(5.0, scale=1),
             norm(-5.0, scale=1),
             ])

dx = 0.01
x = pl.arange(-8, 13, dx)

pl.hist(p.rvs(1500), bins=100, normed=1,alpha=0.25);
pl.plot(x, p.pdf(x), c='r',lw=1);
```

Additionally we create a random variable $f(x)$.

```python
def f(x):
    return 1.0 / (1 + exp(-x))

pl.plot(x, f(x), c='g', alpha=0.8, lw=2)
pl.plot(x, p.pdf(x), c='b', alpha=0.8)

direct = (p.pdf(x) * f(x) * dx).sum()
pl.axvline(direct, c='k', lw=2);
print direct
```

In importance sampling we do not directly sample from $p$, as we would in the Monte carlo method. Instead we sample from a different distribution $q$. 

There are at least two reasons we might  do this: 

  1. It might be more expensive or impossible to sample directly from $p$.
  2. It might be the case that sampling directly from `p` would require more samples to estimate the expectation $\mathbb{E}_{x \sim p}\left[ f(x) \right]$.

Warm up 1: Numerical Integration
================================

We can estimate the integral with a simple numerical approach.

$$
\int_x f(x) p(x) dx \approx \sum_{x} \Delta \cdot f(x) \cdot p(x)
$$

This approach is quite simple to implement, as you can see below, the code looks just like the math!

```python
direct = (p.pdf(x) * f(x) * dx).sum()
print direct
```

Let's plot it

```python
pl.plot(x, f(x), c='g', alpha=0.8)
pl.plot(x, p.pdf(x), c='b', alpha=0.8);
pl.axvline(direct, c='k', lw=2);
```

Warm up 2: Monte Carlo sampling
===============================

Estimates the integral by sampling directly from $p$

$$
\int_x p(x) f(x) dx \approx \frac{1}{n} \sum_{i=1}^n f(x_i)
$$

where $x_1 \cdots x_n$ are i.i.d. samples from $p$.

The code:

```python
ps = p.rvs(1)
mc = f(ps).mean()

pl.plot(x, p.pdf(x), c='k')
pl.axvline(direct, c='b', alpha=0.5, lw=2)
pl.axvline(mc, c='r', alpha=0.5, lw=2);
```

Case 1: Sampling directly from p is expensive
=============================================

```python
N = 1000

q = norm(0.75, scale=4.0)
#q = uniform(loc=-10, scale=20)

pl.plot(x, q.pdf(x), c='r', alpha=0.25)

qs = q.rvs(N)

w = p.pdf(qs) / q.pdf(qs)

im1 = (w * f(qs)).mean()

# estimates
pl.axvline(im1, c='b', alpha=0.5, label='simp')
pl.axvline(direct, color='k', alpha=0.5, label='true')
pl.plot(x, p.pdf(x), c='b', alpha=0.25)

pl.legend()
pl.show()
```

Common method for evaluating the quality of samples is effective sample size.

```python
ess = N / (1 + w.var(ddof=1))
print w.min(), w.max(), w.mean()
print 'ess: %s; N: %s' % (ess, N)
```

Normalized importance sampling
==============================

Divides weights by sum, i.e. w / w.sum()

This allows use to sample from unnormalized densities

Let p' be an unormalized density s.t. p = p'/Zp

It's clear that if w' = p'/q' and w = p/q then w = w'*C where C = Zq/Zp and

$$$
  w'(x) * phi(x) / \sum w'
    = C * w'(x) * phi(x) / sum(C * w')
    = w(x) * phi(x) / sum(w)
$$

$$
\mathbb{E}_q\left[ p'(x) / q'(x) * f(x) \right]
 = Eq[ Zp/Zq p/q * phi]
 = Zp/Zq Eq[ p/q * phi]
 = Zp/Zq Ep[ phi ]
$$

$Z_p/Z_q$ is a mutliplicative bias on actual expectation.

 - Usual method just divides by N instead of sum(w)

   - The cute thing about dividing by sum(x) is that normalization constants
     cancel. This is convenient if the normalizing constants are expensive to compute, which is often the case since they require integration or summation over an exponentially large set.

 - Normalized estimator is biased.

   - all weights are rescaled by sum. If they were all tiny [.1 .1] and their
     sum is tiny =0.2, they are now bigger [0.5 0.5]
