Capped Importance Sampling
==========================

- References:
   - Bottou's slides http://leon.bottou.org/slides/counterfactuals/mlsummit2013.pdf
   - Bottou12counterfactual

- Biased estimator, but bias can be bounded

- Robust confidence interval estimate

```python
R = 5                 # largest weight
M = f(x).max()        # largest value random variable can take

assert ((0 <= f(x)) & (f(x) <= M)).all(), 'f(x) must lie in [0,M]'

print 'clip/capp:', R

capp = w.copy()
capp[w > R] = R

# Using clipped weights eliminates the contribution of samples from poorly
# explored regions -- a large weight a mismatch between p and q.
clip = w.copy()
clip[w > R] = 0.0

im3 = (capp * f(qs)).mean()
im4 = (clip * f(qs)).mean()

# Using these weights we obtain the following (biased) estimator
yhat = (clip * f(qs)).mean()

# the difference between truth and clipped estimates must lie in the following
# interval:

clip_error = M*(1 - clip.mean())    # approximated by the sample weights
print 'clip error bound:', clip_error

delta = 0.05 / 3.0

def bernstein(V, delta, n, a, b):
    assert b >= a
    return sqrt(2 * V * log(2.0/delta) / n) + (b-a)*(7.0/3.0) * log(2.0/delta) / (n-1)

print 'range:', (clip * f(qs)).min(), (clip * f(qs)).max()

# sample variance of clipped weights
V_y = (clip * f(qs)).var(ddof=1)
V_w = clip.var(ddof=1)
eps_R = bernstein(V_y, delta, N, 0, M*R)
eta_R = bernstein(V_w, delta, N, 0, R)

# inner
ci_inner_lb = yhat
ci_inner_ub = yhat + M*(1 - clip.mean() + eta_R)

# outer
ci_outer_lb = yhat - eps_R
ci_outer_ub = yhat + eps_R

# final bound: with probability (1 - 2*delta)
ci_final_lb = yhat - eps_R
ci_final_ub = yhat + M*(1 - clip.mean() + eta_R) + eps_R

print 'interval=[%g, %g] width=%g; prob=%g' % (ci_final_lb, ci_final_ub, ci_final_ub - ci_final_lb, 1-3*delta)
print 'target=%g is=%g nis=%g cap=%g clip=%g' %  (direct, im1, im2, im3, im4)

pl.axvline(im3, c='g', alpha=0.5, label='capp')
pl.axvline(im4, c='y', alpha=0.5, label='clip')
pl.axvline(ci_final_lb, color='k', alpha=0.5, label='lb', linestyle='--')
pl.axvline(ci_final_ub, color='k', alpha=0.5, label='ub', linestyle='--')

pl.plot(x, p.pdf(x), c='b', alpha=0.25)
pl.ylim(0,0.32);
```
